```python
import pandas as pd
import statistics
import os
```

```python
df = pd.read_excel("alternative_tables.xlsx")
```

```python
print(df)
```

```python
rows = {'fomc':[],'a':[],'b':[],'c':[], 'd':[]}
for index,row in df.iterrows():
    if isinstance(row['FOMC'],float)==False: rows['fomc'].append(len(str(row["FOMC"]).split()))
    if isinstance(row['Alternative A'],float)==False: rows['a'].append(len(str(row["Alternative A"]).split())) 
    if isinstance(row['Alternative B'],float)==False: rows['b'].append(len(str(row["Alternative B"]).split()))
    if isinstance(row['Alternative C'],float)==False: rows['c'].append(len(str(row["Alternative C"]).split()))
    if isinstance(row['Alternative D'],float)==False: rows['d'].append((len(str(row["Alternative D"]).split())))
```

```python
print(rows)
```

```python
for alt in rows.keys():
    print(len(rows[alt]))
```
