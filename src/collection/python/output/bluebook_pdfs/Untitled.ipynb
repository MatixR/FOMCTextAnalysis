```python
import pandas as pd
from PyPDF2 import PdfFileReader
import os
import re
```

```python
has_table_df = pd.read_pickle("bluebooks_with_alt_table.pkl")
valid_df = has_table_df[has_table_df.has_table>0]
print(has_table_df.sort_values(by="Date"))
```

```python
df = pd.DataFrame()
for filename in valid_df['Date']:
    full_table = None
    try:
        with open(filename,"rb") as f:
            pdf = PdfFileReader(f)    
            for page in range(pdf.getNumPages()):
                page_text = pdf.getPage(page).extractText()
                if "Alternative Language" in page_text:
                    full_table = page_text
                    break
    except:
        pass
    df = df.append({"Date":filename, "table_text":full_table},ignore_index=True)
```
