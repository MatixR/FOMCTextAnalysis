```python
import pandas as pd
import os
```

```python
df = pd.DataFrame()
counter = 0
for filename in os.listdir():
    with open(filename) as f:
        print(filename)
        present =  f.read().split(":")[1]
        content = []
        for line in present.split("\n"):
            print(line)
            if not line.strip():
                break
            content.append(line)
        df = df.append({"Date":filename,"text":len(content)},ignore_index=True)
    counter += 1
    if counter>20:
        break
print(df)
```
